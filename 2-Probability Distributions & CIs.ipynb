Scores on an examination are assumed to be normally distributed with mean 78 and variance 36.

this is P(X>72)

```python
mu = 78          #mean
sd = np.sqrt(36) #standard deviation
x = 72           #random variable
```

f. If it is known that a student’s score exceeds 72, what is the probability that his or her score exceeds 84?

```python
# P(Y>84|Y>72) = P[Y>84 &(intercept) Y>72] / P(Y>72) = P(Y>84) / P(Y>72)  #conditional probability

(1 - stats.norm.cdf(x=84, loc=78, scale=6)) / (1 - stats.norm.cdf(x=72, loc=78, scale=6))
```

**Confidence Intervals Using the Normal Distribution**


If we’re working with larger samples (n≥30), we can assume that the sampling distribution of the sample mean is normally distributed (thanks to the Central Limit Theorem) and can instead use the norm.interval() function from the scipy.stats library.

```python
sns.get_dataset_names()  #list of dataset names on seaborn
```

```python
#Load "tips" dataset from Seaborn library

tips = sns.load_dataset("tips")
tips.head()
```

```python
#Draw a set of vertical bar plots (based on "total_bill") grouped by "day" variable

sns.barplot(x="day",y="total_bill",data=tips,ci=95);   #ci percentage of confidence interval
```

```python
#Subset a dataframe just including Sunday tips

tipsSun = tips[tips["day"] == "Sun"]
```

```python
tipsSun['total_bill'].mean()
```

```python
tipsSun["total_bill"].std()
```

```python
#Calculate standart error of the mean

sem = tipsSun["total_bill"].std() / np.sqrt(tipsSun.shape[0])
sem
```

```python
tipsSun
```

```python
len(tipsSun)
```

```python
tipsSun.shape  #shape[0]= 76
```

```python
tipsSun["total_bill"].sem()  # sem is described above as a formula.Used as a method there.
```

```python
#Calculate the margin of error (95% ci)  
# m of error= z * standar error of mean
#This number expresses how certain you are that the sample accurately reflects the attitudes of the total population.

moe = 1.96 * sem            # 1.96 is equal to z-score for %95 ci.
moe

```

```python
#Calculate the upper confidence limit

tipsSun["total_bill"].mean() + moe
```

```python
#Calculate the lower confidence limit
tipsSun["total_bill"].mean() - moe
```

```python
#Calculate 95% Confidence Interval around the mean by using scipy.stats

stats.norm.interval(0.95, loc = tipsSun["total_bill"].mean(), scale = tipsSun["total_bill"].sem())
```

**Exercise Large Samples**

The shopping times of n = 64 randomly selected customers at a local supermarket were recorded.

The average and variance of the 64 shopping times were 33  and 256 , respectively.

Estimate μ, the true average shopping time per customer, with a confidence coefficient of 1 − α = .90.

**CI = xbar +- z*s/np.sqrt(n)**

```python
n = 64
xbar = 33
var = 256
ci = 0.90
s = np.sqrt(var)

sem = s/np.sqrt(64)
```

```python
#Using Normal Distribution
stats.norm.interval(0.90, loc = xbar, scale = sem)
```

```python
#Using t Distribution
stats.t.interval(0.90, df=n-1, loc = xbar, scale = sem)
```

**Exercise Small Samples**

A manufacturer of gunpowder has developed a new powder, which was tested in eight shells. The resulting muzzle velocities, in feet per second, were as follows:

3005 2925 2935 2965

2995 3005 2937 2905

Find a 95% confidence interval for the true average velocity μ for shells of this type. Assume that muzzle velocities are approximately normally distributed.

**xbar(sampleMean) +- T ( t for confidence level) * standarderror**

```python
x = [3005, 2925, 2935, 2965, 2995, 3005, 2937, 2905]
```

```python
# Sample size
n = len(x)

# Sample mean
xbar = np.mean(x)

# Sample standard deviation
s = np.std(x,ddof=1)

# Confidence level
cl = 0.95
```

```python
stats.t.interval(cl, df=n-1, loc=xbar, scale=s/np.sqrt(n))
```

```python
#confidence multiplier
t = stats.t.ppf(0.975, df=n-1)
t
```

```python
#standard error of the mean
sem = s/np.sqrt(n)
sem
```

```python
#margin of error
moe = t*sem
moe
```

```python
#lower confidence limit
low_limit = xbar - moe
low_limit
```

```python
#higher confidence limit
high_limit = xbar + moe
high_limit
```
