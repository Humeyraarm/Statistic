```python
# tv_hours vs GPA
tv_hours = [3,5,2,0,5,3,1,4,3,4]
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]
```

```python
tv_hours = np.array(tv_hours)
GPA = np.array(GPA)
```

```python
plt.scatter(tv_hours,GPA);
```

```python
m,b = np.polyfit(tv_hours,GPA,1) # 1 means 1.degree equation-Linear Regression
plt.plot(tv_hours,GPA,"o")
plt.plot(tv_hours,m*tv_hours+b); # converting y=x2 to a linear line y=mx+b
```

```python
# Scipy
```

```python
x = np.array(tv_hours)
y = np.array(GPA)
```

```python
r,p = stats.pearsonr(x,y) #return r and p-value corelation measure method pearsonr assumes both dataset are normally distributed.
                          # p-values lower than 0.05 are usually makes sense
```

```python
stats.spearmanr(x,y) # method spearman r measure assumes that both dataset are not normally distributed.
```

```python
stats.kendalltau(x,y) #Calculate Kendall's tau, a correlation measure for ordinal data.
```

```python
#Numpy
```

```python
np.cov(x,y)     # matris includes measurement cov of itself and with other.
                # gives the direction of the corelation.no other inference.
```

```python
stats.tvar(x)  #This function computes the sample variance of an array of values, 
               #while ignoring values which are outside of given limits
```

```python
stats.tvar(y)
```

```python
np.corrcoef(tv_hours,GPA)  #pearson r ile neredeyse aynı.perason parametrik bir değer.spearman nanparametrik bir ölçüm hesabı.Biz eğer parametrik ve continuous bir değişken kullanıyosak pearson kullanıyoruz.spearman kullnılacak yerleri var akademik.
```

```python
np.corrcoef(x,y)     #nearly same with the pearsonr.
                     #pearsonr is a parametric measure and used for continuous variables.
                     #Spearmanr is nanparametric measure method.Not used usually,used academic.
```
