```python
def region(country):
    if country in ['Mauritius','Congo (Brazzaville)','Ivory Coast','Cameroon','Senegal','Ghana','Niger','Gambia','Benin','Guinea','South Africa','Gabon','Burkina Faso','Mozambique','Nigeria','Mali','Uganda','Liberia','Kenya','Namibia','Chad','Swaziland','Comoros','Ethiopia','Mauritania','Madagascar','Togo','Zambia', 'Sierra Leone','Burundi','Tanzania','Malawi','Lesotho','Botswana','Rwanda','Zimbabwe']:
        return 'Sub-Saharan Africa'
    elif country in ['Finland','Denmark','Switzerland','Iceland','Netherlands','Norway','Sweden','Luxembourg','Austria','Germany','Ireland','United Kingdom','Belgium','France','Malta','Spain','Italy','Cyprus','Portugal','Greece','North Cyprus']:
        return 'Western Europe'
    elif country in ['Costa Rica','Guatemala','Uruguay','Brazil','Mexico','Jamaica','Panama','Chile','El Salvador','Colombia','Nicaragua','Argentina','Honduras','Peru','Ecuador','Bolivia','Paraguay','Dominican Republic','Venezuela','Haiti']:
        return 'Latin America and Caribbean'
    elif country in ['Israel','Bahrain','United Arab Emirates','Saudi Arabia','Kuwait','Libya','Turkey','Morocco','Algeria','Iraq','Iran','Tunisia','Lebanon','Palestinian Territories','Jordan','Egypt','Yemen']:
        return 'Middle East and North Africa'
    elif country in ['Czech Republic','Slovenia','Kosovo','Slovakia','Lithuania','Estonia','Poland','Romania','Serbia','Latvia','Hungary','Croatia','Bosnia and Herzegovina','Montenegro', 'Bulgaria','Albania','North Macedonia']:
        return 'Central and Eastern Europe'
    elif country in ['Uzbekistan','Kazakhstan','Moldova','Kyrgyzstan','Belarus','Russia','Tajikistan','Armenia','Azerbaijan','Turkmenistan','Georgia','Ukraine']:
        return 'Commonwealth of Independent States'
    elif country in ['Singapore','Thailand','Philippines','Vietnam','Malaysia','Indonesia','Laos','Cambodia','Myanmar']:
        return 'Southeast Asia'
    elif country in ['Nepal','Maldives','Bangladesh','Pakistan','Sri Lanka','India','Afghanistan']:
        return 'South Asia'
    elif country in ['Taiwan Province of China','Japan','South Korea','Mongolia','Hong Kong S.A.R. of China','China']:
        return 'East Asia'
    elif country in ['New Zealand', 'Australia', 'Canada', 'United States']:
        return 'North America and ANZ'
```

```python
history_df['Regional indicator'] = history_df['Country name'].apply(region)

history_df

# Regional indicator sütunu oluşturup buna history_df'in country name lerini region fonksiyonuna uygulayarak burdan return olan 
# regionları ona atadık. (fonksiyondaki country parametresine country name değerleri giriyor ve region return oluyor.)
```

```python
# şimdi bu değer aralıklarını fonksiyonda kullanabiliriz. böylece corruption numerik değerlerini
# descriptive olarak tanımalayacağız.

def trust(corrupt):
    if corrupt >= 0.869000:
        return "Low trust in institutions"
    elif corrupt >  0.801:
        return "Lower than normal level of trust"
    elif 0.801 >= corrupt > 0.68875:
        return "About the normal level trust"
    else :
        return "High level of trust"
```

```python
history_df["Trust in institutions"] = history_df["Perceptions of corruption"].apply(trust)

# history_df'te Trust in institutions sütunu oluşturup, içine percentions of corruption değerlerinin 
# trust fonksiyonuna uygulanmış halini atadık.
```
